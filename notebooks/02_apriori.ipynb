```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

import sys
sys.path.append("..")
```

```python
# Reading in dataset of recipes by ingredients
# Reference: https://www.kaggle.com/datasets/alincijov/cooking-ingredients?select=train.csv

# Loading in the dataset
df = pd.read_csv('../data/raw/recipes_by_ingredients.csv', index_col = 0)
df = df.astype(bool)
df
```

```python
# Generating frequent itemsets
frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
```

```python
# Creating association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
#print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
rules.sort_values('lift', ascending = False)
```

```python
# Change frozenset objects to sorted lists to be able to save as parquet
rules["antecedents"] = rules["antecedents"].apply(lambda x: sorted(list(x)))
rules["consequents"] = rules["consequents"].apply(lambda x: sorted(list(x)))
```

```python
def apriori_recommendation(rules: pd.DataFrame, ingredient_query: str):
    
    ingredient_present = rules['antecedents'].apply(lambda fs: ingredient_query in fs).any()
    
    if not ingredient_present:
        print(f"Ingredient '{ingredient_query}' does not appear in any antecedents.")
        return {}
    
    filtered_rules = rules[rules['antecedents'].apply(lambda s: ingredient_query in s) &
                           (rules['lift'] > 1)].copy()
    filtered_rules = filtered_rules.sort_values("lift", ascending = False)
    
    top_recs = {}
    
    for _, row in filtered_rules.iterrows():
        for item in row['consequents']:
            if item not in top_recs:
                top_recs[item] = row['lift']
                
    return top_recs
```

```python
apriori_recommendation(rules, 'tomato paste')
```

```python
apriori_recommendation(rules, 'bulb fennel')
```

```python
possible_ingredients = set()

for _, row in rules.iterrows():
    for ingredient in row['antecedents']:
        possible_ingredients.add(ingredient)

print(f"There are a total of {len(possible_ingredients)} unique ingredients.")
print(possible_ingredients)
```
